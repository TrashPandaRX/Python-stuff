```python
# if you know your data doesnt have column identifier fields (ie a header row) for the first line of data in a file,
# just be sure to say 'header = None' to prevent pandas from automatically assuming the first row is header info.
df8 = pandas.read_csv("supermarkets/supermarkets-semi-colons.txt", sep = ";", header = None)
df8
```

```python
# how to set column names
df8.columns = ["ID2", "Address", "Monkey Town", "Monkey State", "HickZone", "Pet Names", "Goonie Quantity"]
df8
```

```python
# note that set_index does NOT modify existing data, its kinda like a return. if you want to keep using this modified
# version of the data, you need to use the assignment operator to make it persist.
df8.set_index("ID2")
# ps yes ik there are 2 id's but the data i pulled this stuff from already had ID value and i dont feel like removing it yet
```

```python
# or you can use the inplace=True option but you must be cautious as this OVERWRITES the old index, and the old index
# is NOT reassigned, meaning it is lost to the void
df8.set_index("Address", inplace = True)
# i fucked this up, i overused df8 and now its all fucky. redoing what i had prior on "df9"
```

```python
df9 = pandas.read_csv("supermarkets/supermarkets-semi-colons.txt", sep = ";", header = None)
df9.columns = ["ID2", "Address", "Monkey Town", "Monkey State", "HickZone", "Pet Names", "Goonie Quantity"]
df9
```

```python
df9.set_index("Address", inplace = True)
df9
# and it works fine. now back to what the lesson said...
# you must be cautious as this OVERWRITES the old index, and the old index
# is NOT reassigned, meaning it is lost to the void
```

```python
# to avoid losing the index to the void, just add the "drop = False" option
df9.set_index("Monkey Town", inplace = True, drop = False)
df9
# this causes monkey town to be the index, but you keep it in the columns.
```

```python
# however label based access is uncommon. typically you will use "index based access" see data cell labeled "iLOC[]"
df10 = pandas.read_json("supermarkets/supermarkets.json")
df10
```

```python
df10 = df10.set_index("Address")
```

```python
df10.loc[:,"Country"]
```

```python
list(df10.loc[:,"Country"])
```

```python
df10 # just for the sake of referencing
```

```python
df10.loc["3995 23rd St",:]
# PLEASE NOTE!!! slicing with loc[] (the label based one) is upper bound INCLUSIVE
# while iloc (the index based on) is upper bound  EXCLUSIVE (ie [1:4] will only return items 1,2,3).
# please remember this major difference between base loc[] and iloc[] subtle differences in slicing.
```

```python
# LESSON -- iLOC[]
# index-based data access
df10.iloc[2,3:5]
# if a dataframe has an index, it will start counting while excluding the index column.
# which is the "Address" in this case
```

```python
df10.iloc[2,0:4]
```

```python
# LESSON -- df.drop() -- deleting data
# please note, that .drop() is not an inplace operation, so if you want the change to persist, you must assign it
# otherwise the operation's result will be lost
# df.drop("NAME OF COLUMN OR ROW", "0 or 1 -- 0 if you are deleting a row, and 1 if you are deleting a column"),
# but this will soon be deprecated, so that all arguments for drop() except for the argument 'labels'
# will be keyword only. EG do drop("Zoo",0) if you want to get rid of zoo row for now, but soon it will just be
# drop("Zoo")
df10.drop("City", 1)
# df10.drop("City")
```

```python
# LESSON -- Updating & adding new data
df10["Continent"] = df10.shape[0]*["North America"] # .shape[] is the ROW x COLUMNS of the dataframe. ie 5 x 7
# it says 6x7 right now but thats only because i ran this code before the len() checks above
df10
# if you are going to add a new row/column make sure it is the exact dimensions of the existing data
```

```python
# redid this line to reset because i overwrote the following line too many times and it made the continent column ugly
df10["Continent"] = df10.shape[0]*["North America"]
df10["Continent"] = df10["Country"]+ ", " + df10["Continent"]
```

```python
df10
```

```python
# this part is tricky!!!
# adding a new row to a dataframe
# you want to take advantage of the TRANSPOSE method in pandas' dataframes
df10_t = df10.T
# essentially transpose turns columns into rows, and rows into columns
# which effectively means you want a new row OR edit an existing row? then transpose the rows to columns as its easy to make new columns...
# then add a new 'column' with all relevent data, and when you are done with that 'column'
# transpose the dataframe once more back into its original state
# PS dont directly affect your main dataframe if you can avoid it. thats why you make a separate df with _t
```

```python
df10_t
```

```python
df10_t["My Address"] = [7,"My City","My State","My County","My Shop", 51, "My Continent"]
```
